```python
df = data.drop_duplicates() # it removes the duplicate rows in taken dataset
df
```

```python
list(df.columns)
```

fillna() function is using to replace missing values in data fames


```python
df["Category"] = df["Category"].fillna("unknown") # category - filling with "unknown"
```

```python
df["Price"] = df["Price"].fillna(df["Price"].median()) # Price – Fill missing with median (better than mean for price)
```

```python
 #Rating – Fill missing with average rating
df["Rating"] = df["Rating"].fillna(df["Rating"].mean())
```

```python
#Stock – Fill missing with "Not Available"
df["Stock"] = df["Stock"].fillna("Not Available")
```

```python
#Discount – Fill missing with 0
df["Discount"] = df["Discount"].fillna(0)
```

```python
# Convert Stock column to category type
df["Stock"] = df["Stock"].astype("category")
```

```python
# Remove negative prices
df = df[df["Price"] >= 0]
df.head()
```

```python
# Ensure rating is between 0 and 5
df = df[(df["Rating"] >= 0) & (df["Rating"] <= 5)]
df.head()
```

```python
# Ensure discount is not negative
df = df[df["Discount"] >= 0]
df.head()
```

```python
#Reset Index After Cleaning
df = df.reset_index(drop=True)
df.head()
```
