# Write a function that takes the radius of a circle and returns its area.

```python
import math
def circle_radius(radius):
  return math.pi * radius ** 2
radius = int(input("Enter the radius of the circle: "))
print(circle_radius(radius))
```

# **Write a function that accepts three numbers and returns the largest one.**

```python
def largest_number(a, b, c):
  return max(a, b, c)
print(largest_number(10, 20, 30))
```

# **Write a function that takes a string and returns its length without using built-in length functions.**

```python
str1 = input("Enter the string: ")
count = 0
for _ in range(len(str1)):
  count += 1
print(count)
```

# **Write a function that checks whether a number is prime.**

```python
def isPrime(n):
  if n<= 1:
    return False
  for i in range(2, int(n**0.5)+1):
    if n % i ==0:
      return False
  return True
num = int(input("Enter a number:"))
print(isPrime(num))
```

# **Write a recursive function to find the nth Fibonacci number.**

```python
def Fibonacci_number(n):
  if n <= 1:
    return n
  else:
    return Fibonacci_number(n-1) + Fibonacci_number(n-2)
num = int(input("Enter a Number:"))
print(Fibonacci_number(num))
```

# **Write a recursive function to reverse a string.**

```python
def reverse_a_string(str1):
  if len(str1) == 0:
    return str1
  return reverse_a_string(str1[1:])+str1[0]
str1 = input("Enter a String:")
print(reverse_a_string(str1))
```

# **Write a program to find the sum of all numbers from 1 to n using a for loop.**

```python
n = int(input("Enter a number:"))
sum=0
for i in range(1, n+1):
  sum += i
print("sum of all number:",sum)
```

# **Write a program to count how many even numbers are present in a list.**

```python
numbers = [1, 2, 3, 4, 5, 6, 8, 10]
even_count = 0
for num in numbers:
    if num % 2 == 0:
        even_count += 1
print("Even numbers count:", even_count)
```

# **Write a program to print all prime numbers between 1 and 100.**

```python
def isPrime(n):
  if n<= 1:
    return False
  for i in range(2, int(n**0.5)+1):
    if n % i ==0:
      return False
  return True
print("Prime numbers between 1 and 100:")
for num in range(2, 101):
    if isPrime(num):
        print(num, end=' ')
```
